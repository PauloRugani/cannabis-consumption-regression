Data Normality

- Histplot
- Q-Q plot

- Normality tests:
    - Shapiro-Wilk
    - Lilliefors

- Normality test result:

    - Ho = Normal distribution : p > 0.05
    - Ha = distribution != normal : p <= 0.05

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv('../data/processed/2_data_cleaning.csv', sep=',')
df.head(3)
```

```python
for col in df.columns:
  fig, ax = plt.subplots(1, 2, figsize=(14, 6))
  ax[0].hist(df[col], bins=30);
  ax[0].set_title(f'Histogram {col}')
  stats.probplot(df[col], dist='norm', plot=ax[1]);
  ax[1].set_title(f'Q-Q Plot {col}')
```

## Shapiro-Wilk Test

```python
for col in df.columns:
  stat, p = stats.shapiro(df[col])
  if p > 0.05:
    print(f'{col} is normal distribution: {p}')
  else:
    print(f'{col} is not normal distribution: {p}')
```
