# Data Correlation

- Heatmap
- Scatter plot

- Correlation tests:
    - Pearson (Normal distribution)
    - Spearman (Non-normal distribution)
    - Kendall (Non-normal distribution, less data)

- Correlation test result:
    - Ho = Existent Linear Correlation: p > 0,05
    - Ha = Non-existent linear correlation: p <= 0,05

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/2_data_cleaning.csv', sep=',')
df.head(3)
```

```python
sns.pairplot(df, corner=True)
```

```python
plt.figure(figsize=(16, 10))
sns.heatmap(df.corr(method='spearman'), annot=True)
```
