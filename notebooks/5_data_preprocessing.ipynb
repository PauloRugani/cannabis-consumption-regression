# Data preprocessing

- Normalize age and education features

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('../data/processed/2_data_cleaning.csv', sep=',')
df.head(3)
```

```python
std_scaler = StandardScaler()
```

```python
df[['Age', 'Gender' ,'Education']] = std_scaler.fit_transform(df[['Age', 'Gender', 'Education']])
df
```

```python
df.describe()
```

```python
df.to_csv('../data/processed/5_data_preprocessing.csv', index=False)
```
